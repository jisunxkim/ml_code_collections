```python
import pandas as pd
import numpy as np 
```

```python
# load data and impute dirty (duplicate and null)
df1 = pd.read_csv("../datasets/dirty_data/Uncleaned_DS_jobs.csv", index_col='index')

# add dupplicated rows
dup_data = df1.sample(frac=0.2, replace=True)
df1 = pd.concat([df1, dup_data], axis=0)
del dup_data

# add null
cols = ['Salary Estimate', 'Size', 'Type of ownership', 'Revenue']
noise_n = int(df1.size * 0.01)
row_idx = np.random.choice(df1.index, noise_n, replace=True)
col_idx = np.random.choice(cols, noise_n, replace=True)
for r, c in zip(row_idx, col_idx):
    df1.loc[r, c] = None 

df1.head(2)dd
```
