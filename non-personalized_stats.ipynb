```python
#2
# Most popular - rating count
movie_stats = data.iloc[:, 2:].count().sort_values(ascending=False).reset_index()
movie_stats.columns = ['movie', 'rating_count']
movie_stats.head()
```

```python
#3
# % of rating 4 (liked)
temp = data.iloc[:, 2:]
temp = round((temp[temp>=4].count() / temp.count()), 2).reset_index()
temp.columns = ['movie', 'percent_over_4_ratings']
temp.sort_values('percent_over_4_ratings', ascending=False)
# movie_stats = movie_stats.merge(temp, on='movie', how='left')
# movie_stats
```

```python
#4
# top movie from someone who has seen Toy Story:
# counts of raters who have wathced Toy Story
# bayesian  (x+y)/x
# for each movie, calculate the percentage of Toy Story raters who also rated that movie. 

temp = data.iloc[:, 2:]

# x: number of raters who have watched Toy Story
x = temp['1: Toy Story (1995)'].count()

# x + y: number of raters of a movie among who watched Toy Story
xandy = pd.DataFrame(temp[~temp['1: Toy Story (1995)'].isna()].count(), 
                                 columns=['xandy'])

toy_and_others_score = (xandy / x).reset_index()
toy_and_others_score.columns = ['movie', 'xandy_over_x']

# sort by scores and title
toy_and_others_score = toy_and_others_score.sort_values(['xandy_over_x', 'movie'], ascending=False)

toy_and_others_score
```

```python
#5
#Correlation with Toy Story:  Calculate the correlation 
#between the vectors of ratings for Toy Story and each other movie.
#Notice the differences between co-occurrence and correlation; 
#these metrics are showing different types of relationships.
temp = data.iloc[:, 2:]
toystory = data.loc[:, '1: Toy Story (1995)']
corr_with_toysotry = temp.apply(lambda x: x.corr(toystory)).sort_values(ascending=False)
corr_with_toysotry
```

```python
#6
#Mean rating difference by gender:
temp = data\
    .drop('User', axis=1) \
    .groupby('Gender (1 =F, 0=M)').mean() \
    .rename(index={0: 'Male', 1: 'Female'})
temp = temp.transpose()
temp.columns.name = 'Gender'
temp.index.name = 'Movie'
temp.columns = ['Male_avg', 'Female_avg']
temp
```

```python
#overall mean rating (across all ratings) for males and females
male_avg, female_avg = temp.mean(axis=0)
```

```python
female_avg - male_avg
```

```python
# difference between two averages
# Submit the two movies that have the greatest differences 
# (one where men are most above women, and one where women are most above men) 
# along with the differences in average.  Also submit the difference in 
# overall rating averages (female average - male average).
greatest_difference = (temp.Male_avg - temp.Female_avg).agg(['argmin', 'argmax'])
temp.iloc[greatest_difference, :]
```

```python
#7
#Next, compute the % of ratings 4+ separately for males and females.
def over_4_rating(df):
    df.drop('Gender (1 =F, 0=M)', axis=1, inplace=True)
    return df[df>=4].count() / df.count()
    
temp = data.drop('User', axis=1).groupby('Gender (1 =F, 0=M)').apply(lambda x: over_4_rating(x))
temp = temp.rename(index = {1: "Female", 0: 'Male'})

temp =temp.transpose()
temp.index.name = 'Movies'
temp['F-M'] = temp.Female - temp.Male
temp['M-F'] = temp.Male - temp.Female


print("F-M:\n" , temp.iloc[temp['F-M'].argmax(), :])
print("M-F:\n" , temp.iloc[temp['M-F'].argmax(), :])

# first get for each movie the percentage of man / woman that rate >=4. 
# Then the difference between those two percentage. 

# def over_4_rating(df):
#     df.drop('Gender (1 =F, 0=M)', axis=1, inplace=True)
#     return df[df>=4].count()

# temp = data.drop('User', axis=1)
# temp = temp.groupby('Gender (1 =F, 0=M)').apply(lambda x: over_4_rating(x))
# temp = temp.rename(index = {1: "Female", 0: 'Male'})
# temp = temp.transpose()
# temp['F %'] = temp.Female / temp.Male #(temp.Female + temp.Male)
# temp['M %'] = temp.Male / temp.Female #(temp.Female + temp.Male)

# print("Female Max:\n" , temp.iloc[temp['F %'].argmax(), :])
# print("Male Max:\n" , temp.iloc[temp['M %'].argmax(), :])

# temp
```
